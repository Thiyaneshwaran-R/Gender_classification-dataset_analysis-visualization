```python
columns=['forehead_width_cm', 'forehead_height_cm']
for col in columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1

    LOWER_BOUND=Q1-1.5*IQR
    UPPER_BOUND=Q3+1.5*IQR
    outliers=df[(df[col]<LOWER_BOUND)|(df[col]>UPPER_BOUND)]
    print(outliers.shape[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Histogram
```

```python
plt.figure(figsize=(6,5))
sns.histplot(df['forehead_width_cm'],bins=20,kde=True)
plt.title('Histogram for forehead_width')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.histplot(df['forehead_height_cm'],bins=20,kde=True)
plt.title('Histogram for forehead_Height')
plt.show()
```

```python
#Box plot
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df,y='forehead_width_cm')
plt.title('Boxplot')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df,y='forehead_height_cm')
plt.title('Boxplot')
plt.show()
```

```python
#Barchart
```

```python
plt.figure(figsize=(6,5))
Avg_headsize_by_gender=df.groupby('gender')[['forehead_width_cm', 'forehead_height_cm']].mean().reset_index()
Avg_headsize_by_gender.set_index('gender').plot(kind='bar',color=['purple','pink'],width=0.5)
plt.title('Comparision by Gender')
plt.ylabel('Avg headsize by gender')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))
Avg_nosesize_by_gender=df.groupby('gender')[['nose_wide','nose_long']].mean().reset_index()
Avg_nosesize_by_gender.set_index('gender').plot(kind='bar',color=['purple','pink'],width=0.5)
plt.title('Comparision by Gender')
plt.ylabel('Avg Nosesize by gender')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))
Avg_lipssize_by_gender=df.groupby('gender')[['lips_thin','distance_nose_to_lip_long']].mean().reset_index()
Avg_lipssize_by_gender.set_index('gender').plot(kind='bar',color=['purple','pink'],width=0.5)
plt.title('Comparision by Gender')
plt.ylabel('Avg lips size by gender')
plt.tight_layout()
plt.show()
```

```python
#scatterplot
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(data=df,x='forehead_width_cm',y='forehead_height_cm',hue='gender')
plt.title('Scatterplot')
plt.show()
```

```python
#Line chart
```

```python
plt.figure(figsize=(6,5))
df_sorted=df.sort_values(by='forehead_width_cm')
plt.plot(df_sorted['forehead_width_cm'].values,marker='x',linestyle='--')
plt.xlabel('observation')
plt.ylabel('forehead_width_cm')
plt.title('Line chart')
plt.show()
```

```python
#pie chart
```

```python
total_length=df[['forehead_width_cm', 'forehead_height_cm']].sum()
plt.figure(figsize=(6,5))
plt.pie(total_length,labels=total_length.index,autopct='%1.1f%%',startangle=90,colors=['red','yellow'])
plt.title('comparision of forehead width & forehead height')
plt.axis('equal')
plt.show()
```

```python
total_length=df[['nose_wide','nose_long']].sum()
plt.figure(figsize=(6,5))
plt.pie(total_length,labels=total_length.index,autopct='%1.1f%%',startangle=90,colors=['red','yellow'])
plt.title('comparision of nose_wide & nose_long')
plt.axis('equal')
plt.show()
```

```python
total_length=df[['lips_thin','distance_nose_to_lip_long']].sum()
plt.figure(figsize=(6,5))
plt.pie(total_length,labels=total_length.index,autopct='%1.1f%%',startangle=90,colors=['red','yellow'])
plt.title('comparision of thin lips & distance_nose_to_lip_long')
plt.axis('equal')
plt.show()
```
